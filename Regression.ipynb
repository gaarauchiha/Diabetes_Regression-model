```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 

# loading train dataset
data_df = pd.read_csv('train.csv')
data_df.head()
```

```python
y_df = data_df['y']
X_df = data_df.drop(['y'], axis=1)
X_df.head()
```

```python
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_df['Col 1'].values, X_df['Col 2'].values, y_df.values)
ax.set_xlabel('Col 1')
ax.set_ylabel('Col 2')
ax.set_zlabel('y')
```

```python
X = X_df.values
y = y_df.values


X = np.hstack((np.ones((X.shape[0],1)),X))

w = np.dot(np.dot(np.linalg.inv(np.dot(X.T, X)), X.T), y)
w
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
predicted_y = np.dot(X, w)

# training error
print('Mean Squared Error (MSE) for train data set:', metrics.mean_squared_error(y, predicted_y))
```

```python
# Load the training data
train_data = pd.read_csv('train.csv')
X_train = train_data.drop(['id', 'y'], axis=1)
y_train = train_data['y']

# Train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Load the test data
test_data = pd.read_csv('x_test.csv')
X_test = test_data.drop(['id'], axis=1)

# Make predictions on the test data
y_pred = model.predict(X_test)

# Create a DataFrame with the predicted values
prediction_df = pd.DataFrame({'id': test_data['id'], 'y': y_pred})

# Save the predictions to a CSV file
prediction_df.to_csv('y-test.csv', index=False)

y_test = pd.read_csv('y-test.csv')['y']  # Actual test values from test dataset
mse = metrics.mean_squared_error(y_test, y_pred)
print('Mean Squared Error (MSE):', mse)

```

```python
from sklearn.linear_model import LinearRegression

X2 = data_df.drop(['y'], axis=1)
y2 = y_df

model = LinearRegression(fit_intercept=True)

model.fit(X2, y2) # x should be (n_sample, n_feature)
print(model.intercept_, model.coef_)
```

```python
predicted_y = model.predict(X2)

# training error
print('Mean Squared Error (MSE):', metrics.mean_squared_error(y, predicted_y))
```

```python
# with standardization
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

model = LinearRegression(fit_intercept=True)

pipe = make_pipeline(StandardScaler(), model)
pipe.fit(X_test, y_pred)  # apply scaling on training data

print(model.intercept_, model.coef_)
```

```python
# evaluation on training set
predicted_y = pipe.predict(X_test)

# training error
print('Mean Squared Error (MSE):', metrics.mean_squared_error(y_test, predicted_y))
```

```python
from sklearn.linear_model import Ridge
model = Ridge(alpha=1.0)
model.fit(X_train, y_train) 
print(model.intercept_, model.coef_)
predicted_y = model.predict(X_test)
# training error
print('Mean Absolute Error (MAE):', metrics.mean_absolute_error(y_test, predicted_y))
print('Mean Squared Error (MSE):', metrics.mean_squared_error(y_test, predicted_y))
print('Root Mean Squared Error (RMSE):', np.sqrt(metrics.mean_squared_error(y_test, predicted_y)))
```

```python
from sklearn.linear_model import Lasso
model = Lasso(alpha=1.0)
model.fit(X_train, y_train) 
print(model.intercept_, model.coef_)
predicted_y = model.predict(X_test)
# training error
print('Mean Absolute Error (MAE):', metrics.mean_absolute_error(y_test, predicted_y))
print('Mean Squared Error (MSE):', metrics.mean_squared_error(y_test, predicted_y))
print('Root Mean Squared Error (RMSE):', np.sqrt(metrics.mean_squared_error(y_test, predicted_y)))
```
